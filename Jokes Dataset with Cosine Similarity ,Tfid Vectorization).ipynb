```python
def Question_stripping(Inpdata):
    cleanedArticle1=re.sub(r'[?|$|.@#=><|!],Ã‚&*/',r' ',Inpdata)
    cleanedArticle2=re.sub(r'[^a-z A-Z 0-9]',r' ',cleanedArticle1)
    cleanedArticle3= cleanedArticle2.lower()
    cleanedArticle4=re.sub(r'\b\w{1,1}\b', ' ',cleanedArticle3)
    cleanedArticle5=re.sub(r'https?://\S+|www\.\S+',r' ',cleanedArticle4)
    cleanedArticle6=re.sub(r' +', ' ',cleanedArticle5)
    return(cleanedArticle6)
```

```python
Dframe['Ques_Strip']=Dframe['Question'].apply(Question_stripping)
```

```python
def Answer_stripping(Inpdata):
    cleanedArticle1=re.sub(r'[?|$|.@#=><|!]Ã‚&*/',r' ',Inpdata)
    cleanedArticle2=re.sub(r'[^a-z A-Z 0-9]',r' ',cleanedArticle1)
    cleanedArticle3=cleanedArticle2.lower()
    cleanedArticle4=re.sub(r'\b\w{1,1}\b', ' ',cleanedArticle3)
    cleanedArticle5=re.sub(r'https?://\S+|www\.\S+',r' ',cleanedArticle4)
    cleanedArticle6=re.sub(r' +', ' ',cleanedArticle5)
    return(cleanedArticle6)
```

```python
Dframe['Ans_Strip']=Dframe['Answer'].apply(Answer_stripping)
```

```python
Dframe=Dframe.drop(labels=['Question'], axis=1)
```

```python
Dframe=Dframe.drop(labels=['Answer'], axis=1)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
def Question_tokenize(Inpdata):
    tokens = word_tokenize(Inpdata)
    return(tokens)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from sklearn.metrics.pairwise import cosine_similarity
import string
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words=ENGLISH_STOP_WORDS,tokenizer=Question_tokenize,ngram_range=(1, 2))
tfidf_matrix = tfidf_vectorizer.fit_transform(Dframe['Ques_Strip'])
```

```python
def ask_question(question):
    querry=tfidf_vectorizer.transform([question])
    similarity = cosine_similarity(querry, tfidf_matrix)
    max_similarity = np.argmax(similarity, axis=None)
    print('Your question:', question)
    print('Closest question found:', Dframe.iloc[max_similarity]['Ques_Strip'])
    print('Similarity:',similarity[0, max_similarity])
    print('Answer:',Dframe.iloc[max_similarity]['Ans_Strip'])
```
